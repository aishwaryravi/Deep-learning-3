```python
import torch
import torch.nn as nn

import matplotlib.pyplot as plt
torch.manual_seed(2)
```

```python
z = torch.arange(-10, 10, 0.1).view(-1,1)
```

```python
# create a sigmoid object

sig = nn.Sigmoid()
```

```python
yhat = sig(z)
```

```python
# plot the results

plt.plot(z.detach().numpy(), yhat.detach().numpy())
plt.xlabel('z')
plt.ylabel('yhat')
```

```python
# use build in function to plot the result

yhat = torch.sigmoid(z)
plt.plot(z.numpy(), yhat.numpy())

plt.show()
```

```python
TANH = nn.Tanh()
```

```python
yhat = TANH(z)
plt.plot(z.numpy(), yhat.numpy())

plt.show()
```

```python
yhat = torch.tanh(z)
plt.plot(z.numpy(), yhat.numpy())

plt.show()
```

```python
# create a relu object and make a prediction

RELU = nn.ReLU()
yhat = RELU(z)
plt.plot(z.numpy(), yhat.numpy())
```

```python
yhat = torch.relu(z)
plt.plot(z.numpy(), yhat.numpy())

plt.show()
```

```python
# compare activation functions

x = torch.arange(-2, 2, 0.1).view(-1,1)

plt.plot(x.numpy(), torch.sigmoid(x).numpy(),label ='sigmoid')
plt.plot(x.numpy(), torch.tanh(x).numpy(), label = 'tanh')
plt.plot(x.numpy(), torch.relu(x).numpy(), label = 'relu')
plt.legend()
plt.show()
```
